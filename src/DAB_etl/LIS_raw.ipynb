```python
from pyspark.sql.functions import *

load_path = "abfss://raw@hisadls.dfs.core.windows.net/LISAPI"
# checkpoint_path = "abfss://checkpoint@hisadls.dfs.core.windows.net/LISAPI"

raw_df = spark.read.format("text")\
    .load(load_path)
```

```python
bronze_df = raw_df.withColumn("ingested_at", current_timestamp())\
    .withColumn("input_File_name", col("_metadata.file_path"))

bronze_df.display()
```

```python
bronze_df.write.format("delta")\
    .mode("append")\
        .saveAsTable("surakshadb.bronze_stg.lis_data_raw")
```
